# Install MongoDB

```python
%%bash
apt-get install gnupg curl
curl -fsSL https://www.mongodb.org/static/pgp/server-7.0.asc | \
   sudo gpg -o /usr/share/keyrings/mongodb-server-7.0.gpg \
   --dearmor
echo "deb [ arch=amd64,arm64 signed-by=/usr/share/keyrings/mongodb-server-7.0.gpg ] https://repo.mongodb.org/apt/ubuntu jammy/mongodb-org/7.0 multiverse" | sudo tee /etc/apt/sources.list.d/mongodb-org-7.0.list
apt-get update
apt-get install -y mongodb-org
```

```python
#8
dish_id = 1

pipeline = [
    {"$match": {"_id": dish_id}},
    {"$unwind": "$escandall"},
    {
        "$lookup": {
            "from": "proveidors",
            "let": {"escandall_nom": "$escandall.nom", "escandall_id": "$escandall.ingredient_id"},
            "pipeline": [
                {"$unwind": "$cataleg_preus"},
                {
                    "$match": {
                        "$expr": {
                            "$and": [
                                {"$eq": ["$cataleg_preus.nom", "$$escandall_nom"]},
                                {"$eq": ["$cataleg_preus.ingredient_id", "$$escandall_id"]}
                            ]
                        }
                    }
                },
                {"$project": {"_id": 0, "cost_per_unit": "$cataleg_preus.cost"}},
                {"$limit": 1} # Agafar només un cost si un ingredient és subministrat per múltiples proveïdors
            ],
            "as": "matched_ingredient_info"
        }
    },
    {"$unwind": "$matched_ingredient_info"},
    {"$group": {
        "_id": "$nom",
        "total_cost": {
            "$sum": {"$multiply": ["$escandall.quantitat", "$matched_ingredient_info.cost_per_unit"]
            }
        }
    }},
    {"$project": {
        "_id": 0,
        "nom_plat": "$_id",
        "cost_total_ingredients": "$total_cost"
    }}
]

result = list(db.plats.aggregate(pipeline))
print(result)
```

```python
#10
pipeline_total_compres = [
    {"$unwind": "$comandes_proveiment_fetes"},
    {
        "$addFields": {
            "matched_item": {
                "$arrayElemAt": [
                    {
                        "$filter": {
                            "input": "$cataleg_preus",
                            "as": "item",
                            "cond": {
                                "$and": [
                                    {"$eq": ["$$item.ingredient_id", "$comandes_proveiment_fetes.ingredient_id"]},
                                    {"$eq": ["$$item.nom", "$comandes_proveiment_fetes.nom"]}
                                ]
                            }
                        }
                    },
                    0
                ]
            }
        }
    },
    {
        "$addFields": {
            "comandes_proveiment_fetes.cost_per_unit": "$matched_item.cost"
        }
    },
    {
        "$group": {
            "_id": "$_id",
            "nom_proveidor": {"$first": "$nom"},
            "import_acumulat": {
                "$sum": {
                    "$multiply": [
                        "$comandes_proveiment_fetes.quantitat",
                        "$comandes_proveiment_fetes.cost_per_unit"
                    ]
                }
            }
        }
    },
    {
        "$project": {
            "_id": 0,
            "nom_proveidor": "$nom_proveidor",
            "import_acumulat": {"$round": ["$import_acumulat", 2]}
        }
    }
]

results_total_compres = list(db.proveidors.aggregate(pipeline_total_compres))

if results_total_compres:
    print("Import acumulat de les compres efectuades a cada proveïdor:")
    for result in results_total_compres:
        print(f"  {result['nom_proveidor']}: {result['import_acumulat']:.2f}€")
else:
    print("No s'ha pogut calcular l'import acumulat per a cap proveïdor.")
```

```python
#12
print("Proveïdor principal per a cada ingredient (el que més unitats ha subministrat):")

pipeline_main_supplier = [
    {"$unwind": "$cataleg_preus"},
    {"$unwind": "$comandes_proveiment_fetes"},
    {"$match": {
        "$expr": {
            "$eq": [
                "$cataleg_preus.ingredient_id",
                "$comandes_proveiment_fetes.ingredient_id"
            ]
        }
    }},
    {"$group": {
        "_id": {
            "ingredient_id": "$cataleg_preus.ingredient_id",
            "ingredient_nom": "$cataleg_preus.nom",
            "proveidor_id": "$_id",
            "proveidor_nom": "$nom"
        },
        "total_quantitat_subministrada": {"$sum": "$comandes_proveiment_fetes.quantitat"}
    }},
    {"$group": {
        "_id": "$_id.ingredient_id",
        "ingredient_nom": {"$first": "$_id.ingredient_nom"},
        "proveidor_principal": {"$max": {
            "quantitat": "$total_quantitat_subministrada",
            "id": "$_id.proveidor_id",
            "nom": "$_id.proveidor_nom"
        }}
    }},
    {"$project": {
        "_id": 0,
        "ingredient_id": "$_id",
        "ingredient_nom": "$ingredient_nom",
        "proveidor_principal_id": "$proveidor_principal.id",
        "proveidor_principal_nom": "$proveidor_principal.nom",
        "quantitat_total_subministrada": "$proveidor_principal.quantitat"
    }}
]

main_suppliers = list(db.proveidors.aggregate(pipeline_main_supplier))

for supplier_info in main_suppliers:
    print(f"  - Ingredient ID: {supplier_info['ingredient_id']}, Nom: {supplier_info['ingredient_nom']}\n    Proveïdor Principal: {supplier_info['proveidor_principal_nom']} (ID: {supplier_info['proveidor_principal_id']}), Quantitat Total Subministrada: {supplier_info['quantitat_total_subministrada']}")
```

```python
#13
print("Ingredients subministrats per més d'un proveïdor:")

pipeline_multiple_providers = [
    {"$unwind": "$cataleg_preus"},
    {"$group": {
        "_id": {
            "ingredient_id": "$cataleg_preus.ingredient_id",
            "ingredient_nom": "$cataleg_preus.nom"
        },
        "proveidors_unics": {"$addToSet": "$_id"}
    }},
    {"$project": {
        "_id": 0,
        "ingredient_id": "$_id.ingredient_id",
        "ingredient_nom": "$_id.ingredient_nom",
        "num_proveidors": {"$size": "$proveidors_unics"}
    }},
    {"$match": {
        "num_proveidors": {"$gt": 1}
    }}
]

ingredients_multiple_providers = list(db.proveidors.aggregate(pipeline_multiple_providers))

if ingredients_multiple_providers:
    for ingredient in ingredients_multiple_providers:
        print(f"  - Ingredient ID: {ingredient['ingredient_id']}, Nom: {ingredient['ingredient_nom']}, Proveïdors: {ingredient['num_proveidors']}")
else:
    print("No s'han trobat ingredients subministrats per més d'un proveïdor.")
```
