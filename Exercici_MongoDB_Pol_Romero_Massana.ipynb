**EXERCICI 1**

```python
from pymongo import MongoClient
client = MongoClient()
meteo = client.meteo
weather = meteo.weather
```

```python
import datetime
import random
mostra = {
    "date": "2023-10-20",
    "temperature": 25.0,
    "humidity": 60.0,
    "wind_speed": 10.0,
}
weather.drop()
weather.insert_one(mostra)

dades_aleatories = []
date = datetime.datetime(2024, 1, 1)
for i in range(365):
    next_date = date + datetime.timedelta(days=i)
    registre = {
        "date": next_date.strftime("%Y-%m-%d"),
        "temperature": round(random.uniform(20.0, 30.0), 1),
        "humidity": round(random.uniform(50.0, 80.0), 1),
        "wind_speed": round(random.uniform(5.0, 15.0), 1),
    }
    dades_aleatories.append(registre)
weather.insert_many(dades_aleatories)


import pandas as pd
days = weather.find()
list_days = list(days)
df = pd.DataFrame(list_days)
df

```

```python
#5
days1 = weather.find({ "temperature": { "$gt": 25 } })
df1 = pd.DataFrame(list(days1))
df1
```

```python
days2 = weather.find({ "wind_speed": { "$gt": 10 } })
df2 = pd.DataFrame(list(days2))
df2
```

```python
days3 = weather.find({ "humidity": { "$lt": 70 } })
df3 = pd.DataFrame(list(days3))
df3
```

```python
days4 = weather.find({ "temperature": { "$gt": 20 }, "temperature": { "$lt": 25 }})
df4 = pd.DataFrame(list(days4))
df4
```

```python
days5 = weather.find({ "wind_speed": { "$gt": 10 }, "humidity": { "$lt": 60 }})
df5 = pd.DataFrame(list(days5))
df5
```

```python
#10
import json

def prepare_df_for_json(df_input):
    df_copy = df_input.copy()
    if '_id' in df_copy.columns:
        df_copy['_id'] = df_copy['_id'].astype(str)
    return df_copy.to_dict(orient='records')

all_results = {
    "days1_temp_gt_25": prepare_df_for_json(df1),
    "days2_wind_gt_10": prepare_df_for_json(df2),
    "days3_humidity_lt_70": prepare_df_for_json(df3),
    "days4_temp_20_to_25": prepare_df_for_json(df4),
    "days5_wind_gt_10_humidity_lt_60": prepare_df_for_json(df5)
}

# Write results to a JSON file
with open('weather_results.json', 'w') as f:
    json.dump(all_results, f, indent=4)
```

```python
#11
import random

for doc in weather.find():
    random_rain = round(random.uniform(0.0, 10.0), 1)
    weather.update_one(
        {"_id": doc["_id"]},
        {"$set": {"rain": random_rain}}
    )
df = pd.DataFrame(list(weather.find()))
df
```

**EXERCICI 2**

```python
from pymongo import MongoClient
client = MongoClient()
cattle = client.cattle
```

```python
import json
import pandas as pd

df_ganado = pd.read_json('/content/sample_data/censo-ganado.json')

census = cattle.census

cattle.ganado.insert_many(df_ganado.to_dict('records'))
```

```python
# 3
cattle_gt_1000 = cattle.ganado.find({"Número de reses": {"$gt": 1000}})

df_cattle_gt_1000 = pd.DataFrame(list(cattle_gt_1000))

df_cattle_gt_1000
```

```python
#4
cattle_filtered = cattle.ganado.find({
    "Número de reses": {"$gt": 100},
    "Número de explotaciones": {"$gt": 10}
})

df_cattle_filtered = pd.DataFrame(list(cattle_filtered))

df_cattle_filtered
```

```python
#5
cattle_soba_ovino = cattle.ganado.find({
    "Especie": "Ovino",
    "Municipio": "39083 - Soba"
})


df_soba_ovino = pd.DataFrame(list(cattle_soba_ovino))

df_soba_ovino
```

```python
#6
import plotly.express as px


fig = px.bar(df_soba_ovino.groupby('Año')['Número de reses'].sum().reset_index(),
             x="Año", y="Número de reses",
             color = "Número de reses",
             title="Número Total de Reses Ovinas por Año en Soba")
fig.show()
```

```python
#7
for doc in cattle.ganado.find():
    municipio_full = doc.get('Municipio')
    if municipio_full and ' - ' in municipio_full:
        municipality_code, municipality_name = municipio_full.split(' - ', 1)
        cattle.ganado.update_one(
            {'_id': doc['_id']},
            {
                '$set': {
                    'municipality_code': municipality_code.strip(),
                    'municipality_name': municipality_name.strip()
                },
                '$unset': {'Municipio': ''} # Remove the old field
            }
        )

df = pd.DataFrame(list(cattle.ganado.find()))
df
```

```python
#8
import json

data = list(cattle.ganado.find({}))

for doc in data:
    doc['_id'] = str(doc['_id'])

with open('cattle_results.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4)
```
